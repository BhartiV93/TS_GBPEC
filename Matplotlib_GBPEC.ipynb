```python
import numpy as np

x = np.arange(0,10,0.2)

y = np.sin(x)

plt.figure(figsize=(3,3))


plt.xlabel("x")

plt.ylabel("sin(x)")

plt.title("My sine function")

plt.plot(x,y)

plt.show()
```

```python
x = np.array([1,2 , 3, 4, 5])

y = np.array([7,4,8,5,7])

plt.plot(x,y,'green',marker='o',ms=15,mfc="pink",mec="green",ls='solid')

plt.grid(True,c="k")

plt.show()
```

```python

x1 = np.array([1,2,3,4])
y1 =np.array([5,6,7,8])

x2 = np.array([1,2,3,4])
y2 = np.array([3,5,7,8])

plt.plot(x1,y1,"green",label="lineone",linewidth="3",marker='o',ms=10)

plt.plot(x2,y2,"b",label="linetwo",linewidth="2",marker="*",ms=10)

plt.grid(True,color="k")

plt.legend(loc="upper right")

plt.title("my grid plot")

plt.xlabel("x")

plt.ylabel("y")

plt.show()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(6, 5))



x = [0, 3, 5]

y = [1, 4, 3.5]

#label_x = 1
#label_y = 4
#arrow_x = 3
#arrow_y = 4

arrow_properties = dict(
    facecolor="black", width=2,
    headwidth=5)

plt.annotate(
    "maximum", xy=(3,4),
    xytext=(1,4),
    arrowprops=arrow_properties)




plt.plot(x, y)

plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt


fig = plt.figure(figsize=(5,5))

# syntax for 3-D projection

ax = plt.axes(projection ='3d')

# defining all 3 axes

z = np.linspace(0, 1, 100)

x = z * np.sin(25 * z)

y = z * np.cos(25 * z)

# plotting
ax.plot3D(x, y, z, 'green')

ax.set_title('3D line plot')

plt.show()
```

```python
import matplotlib.pyplot as plt

#define subplots
fig, ax = plt.subplots(3, 3, figsize=(5,5))

fig.tight_layout()

#define data
x = [1, 2, 3]
y = [7, 13, 24]

#create subplots
ax[0,0].plot(x, y, color='red')
ax[0,1].plot(x, y, color='blue')
ax[0,2].plot(x, y, color='blue')
ax[1,0].plot(x, y, color='green')
ax[1,1].plot(x, y, color='purple')

plt.show()
```

```python
def org_plot(ax):
    ax.plot([1, 2])
    
    ax.set_xlabel('x-label')
    
    ax.set_ylabel('y-label')
    
    ax.set_title('Title')
```

```python
fig, ax = plt.subplots()
org_plot(ax)
plt.tight_layout()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
org_plot(ax1)
org_plot(ax2)
org_plot(ax3)
org_plot(ax4)
plt.tight_layout()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
org_plot(ax1)
org_plot(ax2)
plt.tight_layout()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
org_plot(ax1)
org_plot(ax2)
plt.tight_layout()
```

```python
fig = plt.figure()

ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(223)
ax3 = plt.subplot(122)

org_plot(ax1)
org_plot(ax2)
org_plot(ax3)

plt.tight_layout()
```

```python
fig = plt.figure()

ax1 = plt.subplot2grid((3, 3), (0, 0))
ax2 = plt.subplot2grid((3, 3), (0, 1), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 0), colspan=2, rowspan=2)
ax4 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)

org_plot(ax1)
org_plot(ax2)
org_plot(ax3)
org_plot(ax4)

plt.tight_layout()

plt.show()
```
